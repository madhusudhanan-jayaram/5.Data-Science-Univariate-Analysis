```python
# Initialize an empty DataFrame 'descriptive' with summary-statistic rows and quantitative columns
descriptive = pd.DataFrame(index=["Mean", "Median", "Mode", "Q1:25%", "Q2:50%", "Q3:75%",  "P99:99%", "Q4:100%", "IQR","1.5rule","Lesser","Greater","Min","Max"], columns=quan)

# Loop through each quantitative column and assign stats using .loc to avoid chained assignment
for columnName in quan:
    # Compute the average and store it at row "Mean", column 'columnName'
    descriptive.loc["Mean", columnName]   = dataset[columnName].mean()
    # Compute the median and store it at row "Median", column 'columnName'
    descriptive.loc["Median", columnName] = dataset[columnName].median()
    # Compute the mode (take the first most frequent value) and store it at row "Mode"
    descriptive.loc["Mode", columnName]   = dataset[columnName].mode()[0]
    # 25th percentile (first quartile)
    descriptive.loc["Q1:25%",  columnName] = dataset[columnName].quantile(0.25)
    # 50th percentile (second quartile, same as median)
    descriptive.loc["Q2:50%",  columnName] = dataset[columnName].quantile(0.50)
    # 75th percentile (third quartile)
    descriptive.loc["Q3:75%",  columnName] = dataset[columnName].quantile(0.75)
    descriptive.loc["P99:99%",  columnName] = dataset[columnName].quantile(0.99)  # 99th percentile
    # 100th percentile (max value)
    descriptive.loc["Q4:100%", columnName] = dataset[columnName].max()
    # Compute the interquartile range (IQR) for each quantitative column
    descriptive.loc["IQR",      columnName] = (
        descriptive.loc["Q3:75%", columnName]  # 75th percentile
      - descriptive.loc["Q1:25%", columnName]  # 25th percentile
    )
    
    # Compute the 1.5×IQR rule threshold (commonly used to detect outliers)
    descriptive.loc["1.5rule", columnName] = 1.5 * descriptive.loc["IQR", columnName]
    
    # Calculate the lower “fence”: Q1 − 1.5×IQR
    descriptive.loc["Lesser",  columnName] = (
        descriptive.loc["Q1:25%", columnName]
      - descriptive.loc["1.5rule", columnName]
    )
    
    # Calculate the upper “fence”: Q3 + 1.5×IQR
    descriptive.loc["Greater", columnName] = (
        descriptive.loc["Q3:75%", columnName]
      + descriptive.loc["1.5rule", columnName]
    )
    
    # Record the true minimum and maximum observed values in the dataset
    descriptive.loc["Min",     columnName] = dataset[columnName].min()  # Smallest value
    descriptive.loc["Max",     columnName] = dataset[columnName].max()  # Largest value
```

```python
# Loop through columns with low outliers and cap them at the Lesser bound
for columnName in lesser:
    # Replace values below the Lesser bound with the Lesser bound (low outlier capping)
    dataset[columnName][dataset[columnName] < descriptive[columnName]["Lesser"]] = descriptive[columnName]["Lesser"]

# Loop through columns with high outliers and cap them at the Greater bound
for columnName in greater:
    # Replace values above the Greater bound with the Greater bound (high outlier capping)
    dataset[columnName][dataset[columnName] > descriptive[columnName]["Greater"]] = descriptive[columnName]["Greater"]
```

```python
# Create a descriptive statistics DataFrame with required metrics as rows and quantitative columns as columns
descriptive = pd.DataFrame(
    index=["Mean", "Median", "Mode", "Q1:25%", "Q2:50%", "Q3:75%", "99%", "Q4:100%", 
           "IQR", "1.5rule", "Lesser", "Greater", "Min", "Max"], 
    columns=quan
)

# Loop through each quantitative column in the dataset
for columnName in quan:
    # Central tendency measures
    descriptive[columnName]["Mean"] = dataset[columnName].mean()     # Average
    descriptive[columnName]["Median"] = dataset[columnName].median() # Middle value
    descriptive[columnName]["Mode"] = dataset[columnName].mode()[0]  # Most frequent value

    # Quartiles (Q1, Q2, Q3)
    descriptive[columnName]["Q1:25%"] = dataset.describe()[columnName]["25%"] # First quartile
    descriptive[columnName]["Q2:50%"] = dataset.describe()[columnName]["50%"] # Second quartile (Median)
    descriptive[columnName]["Q3:75%"] = dataset.describe()[columnName]["75%"] # Third quartile

    # 99th percentile and max value (Q4:100%)
    descriptive[columnName]["99%"] = np.percentile(dataset[columnName], 99)    # 99th percentile
    descriptive[columnName]["Q4:100%"] = dataset.describe()[columnName]["max"] # Maximum value

    # IQR and 1.5×IQR rule
    descriptive[columnName]["IQR"] = descriptive[columnName]["Q3:75%"] - descriptive[columnName]["Q1:25%"]
    descriptive[columnName]["1.5rule"] = 1.5 * descriptive[columnName]["IQR"]

    # Outlier bounds (Lesser = lower bound, Greater = upper bound)
    descriptive[columnName]["Lesser"] = descriptive[columnName]["Q1:25%"] - descriptive[columnName]["1.5rule"]
    descriptive[columnName]["Greater"] = descriptive[columnName]["Q3:75%"] + descriptive[columnName]["1.5rule"]

    # Min and Max values from dataset
    descriptive[columnName]["Min"] = dataset[columnName].min()
    descriptive[columnName]["Max"] = dataset[columnName].max()
```
